```python
import requests
import lxml
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
```

```python
negativas = "Negative.txt"
positivas ="Positive.txt"
```

```python
#leemos el archivo xml 

print("leyendo el archivo")
url="https://itunes.apple.com/us/rss/customerreviews/page=1/id=284882215/sortBy=mostrecent/xml"
datos_xml =requests.get(url).content


#parsear el archivo xml 

print("parseando y cargando beatifulsoup")
datos_bs= BeautifulSoup(datos_xml, "lxml")

```

```python
#almacenar comentarios en una lista e imprimirlos

lista_comentarios =[]
lista_comentarios= datos_bs.find_all("content")
#puntuacion = '()[],.""¬ø?:;_-¬°!‚Äú‚Äù0123456789¬∞|¬¨#$%/=\*¬®¬®{+}'
puntuacion = [ ',',  ':', ';', '.',  '¬ø', '!', '¬°', '-', '"', '(', ')', '[', ']', '"']
limpio =""
lista_de_tokens=[]
comentarios=[]

contador=0
for index, elemento in enumerate(lista_comentarios):
    contador= contador+1
    
    if(index%2==0):
        comentarios.append(elemento.text)
        #print(elemento.text)
        for caracter in elemento.text:            
            if(caracter not in puntuacion):
                limpio = limpio + caracter           
                
        #print("\nComentario"+str(contador)+":\n"+limpio)
              
    tokens = word_tokenize(limpio)
    tokens = [x.lower() for x in tokens] 
    
    if(tokens):  
        #print(tokens)      
        lista_de_tokens.append(tokens)           
    limpio =""

print("\ncomentarios totales:"+str(len(lista_de_tokens)))
#print(lista_de_tokens)
#print(comentarios)
```

```python
#Lista de palabras positivas

with open(positivas, "r", encoding="utf-8") as archivo:
    lista_positivas = list (map(str.rstrip, archivo))
    print('\nPALABRAS POSITIVAS:\n', lista_positivas)

```

```python
#Lista de palabras negativas

with open(negativas, "r", encoding="utf-8") as archivo:
    lista_negativas = list (map(str.rstrip, archivo))
print('\nPALABRAS NEGATIVAS:\n', lista_negativas)
```

```python

#Contar palabras positivas en el texto
print('\nPALABRAS positivas:\n')

for comentario in lista_de_tokens:
    contadorLeal = 0
    contadorBellaco = 0
    for palabra in comentario:
        if palabra in lista_positivas:
            contadorLeal+=1
            print(":)"+palabra)
        if palabra in lista_negativas:
            print(":("+palabra)
            contadorBellaco+=1
    print(comentario)
    print('total de palabras positivas:', contadorLeal)
    print('total de palabras negativas:', contadorBellaco,'\n')

```
